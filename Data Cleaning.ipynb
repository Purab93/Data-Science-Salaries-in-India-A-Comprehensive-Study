```python
def mean_salary(salary):
    dummy = salary.replace('₹','').replace(' ','').replace(',','').replace('K','000').replace('/hr','').replace('/mo','').replace('£','')
    dummy = list(map(float,dummy.split('-')))
    dummy = round(sum(dummy)/len(dummy))
    # this is not exactly advisable to perform 
    #here I am asuming that the salary given is lesser than 100000 then it must be given in monthly format
    # as I have mentioned the salary data is in txt format, checking for the presence of /mo is difficult. hence I am taking this approach
    if dummy < 100000: 
        dummy = dummy*12
    return dummy
        
```

```python
def is_DS(name):
    name=name.lower()
    if ('data science' in name)or('scientist' in name)or('analyst'in name):
        return True
    else:
        return False
```

#### Logic : 
I am trying a complicated Logic here : 
I am checking the corresponding salary listing for each job postion. 
If out of them there is a job title with the term data science/scientist/analyst i am taking the avarage of those salaries 
If the name dosent have those terms in the job openings then I will be taking the average of the remaining.

__I shall elaborate on the reasons for these steps along this file__

Cleaning the salary data is complicated as the data is not in string format, rather in text format, I suspect it is because if the rupee sign

```python
salary_list=[]
counter = 0 
# running through all the job listing
for i in jobs.index.values:
    
    # all the corresponding salary listed for i-th job 
    sal_ds = []
    sal_other = []
    while(salary.loc[counter]['Company Name'] == jobs.loc[i]['Company Name']):
        
        #converting salary to interger format and finding it mean
        try: 
            mean_sal = mean_salary(salary.loc[counter]['Salary'])
        except:
            mean_sal = -1
        
        #checking if there is a data science salary available, storing the value
        if is_DS(salary.loc[counter]['Job Title']):
            sal_ds.append(mean_sal)
        else:
            sal_other.append(mean_sal)
        counter+=1
        
        if(counter>=salary.shape[0]): 
            print('Done')
            break
            
#     print("DS - ",sal_ds,"Other - " , sal_other) #Debugging
    
    #checking if the company has any salary for data science related role and storing the avg. salary of the roles
    #if no data scince related salary is posted then we are going to store the average of the other salaries 
    if sum(sal_ds)>0:
        salary_list.append(sum(sal_ds)/len(sal_ds))    
    elif sum(sal_other)>0:
        salary_list.append(sum(sal_other)/len(sal_other))
    else:
        salary_list.append(-1)
    
    
```

#### Explanation:

The reason i followed this complicatedd procedure instead of grouping the data according to the company and finding the mean of the the salaries is because, every job opeing had corresponding salary details. The same company will have openings in different locations/different titles. And that would lead to in-accurate data

```python
jobs['Salary']=salary_list
```
